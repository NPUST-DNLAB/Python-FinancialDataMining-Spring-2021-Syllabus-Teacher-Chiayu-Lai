```python
import pandas as pd
import statsmodels.api as sm
```

<font size=6>**head<p>2014-01<p>2498&Bank<p>CAMP**</font>

```python
TPE2498 = pd.read_csv('2498.csv', parse_dates=True, index_col='Date')
TSE = pd.read_csv('^TWII.csv', parse_dates=True, index_col='Date')

daily_prices = pd.concat([TPE2498['Close'], TSE['Close']], axis=1)
daily_prices.columns = ['TPE2498', 'TSE']

print(daily_prices.head())

daily_returns = daily_prices.pct_change(1)
clean_daily_returns = daily_returns.dropna(axis=0) 
print(clean_daily_returns.head())
```

```python
# split dependent and independent variable
X = clean_daily_returns['TPE2498']
y = clean_daily_returns['TSE']

# Add a constant to the independent value
X1 = sm.add_constant(X)

# make regression model 
model = sm.OLS(y, X1)

# fit model and print results
results = model.fit()
print(results.summary())
```

<font size=6>**三因子模型**</font>

```python
stock = pd.read_csv('2498.csv', parse_dates=True, index_col='Date')
data = pd.read_csv('factors.csv', index_col='YY/MM/DD')
stock.head(n=3)
UniPresident = stock['ROI']
```

```python
regThrFac = sm.OLS(UniPresident,sm.add_constant(data.iloc[:,2:5]))
result = regThrFac.fit()
print(result.summary())
print("\n\n\n\n\n\n\n")
print(result.params)
```
